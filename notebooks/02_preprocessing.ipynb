Merges datasets, handles missing values, clips outliers, and applies flood thresholds.

```python
import sys
sys.path.append('..')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from src.preprocessing import (
    merge_datasets, preprocess_pipeline,
    calculate_flood_thresholds, apply_future_flood_flags
)
```

```python
# ── Merge raw sources ─────────────────────────────────────────
master = merge_datasets(
    rain_path='../data/CHIRPS_monthly_data_2010_2024.csv',
    tmax_path='../data/TerraClimate_Tmax_monthly_2010_2025.csv',
    static_path='../data/static_features.csv',
    flood_2010_2018_path='../data/Pakistan_Flood_Flags_2010_2018.csv'
)
master.to_csv('../data/master_with_floodflags_raw.csv', index=False)
print(master.shape)
master.head()
```

```python
# ── Run full preprocessing pipeline ──────────────────────────
df_clean = preprocess_pipeline('../data/master_with_floodflags_raw.csv')
df_clean.to_csv('../data/master_with_floodflags.csv', index=False)
print('Flood rate:', df_clean['Flood_Flag'].mean()*100, '%')
df_clean.describe()
```

```python
# ── Visualise class balance ───────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Flood flag distribution
df_clean['Flood_Flag'].value_counts().plot(kind='bar', ax=axes[0], color=['steelblue','tomato'])
axes[0].set_title('Class Distribution')
axes[0].set_xticklabels(['No Flood', 'Flood'], rotation=0)

# Floods by month
df_clean[df_clean['Flood_Flag']==1]['month'].value_counts().sort_index().plot(
    kind='bar', ax=axes[1], color='tomato')
axes[1].set_title('Flood Events by Month')
axes[1].set_xlabel('Month')
plt.tight_layout()
plt.savefig('../results/class_balance.png', dpi=150)
plt.show()
```
